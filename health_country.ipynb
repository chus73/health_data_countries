### Preselección de variables

Vamos a utilizar una técnica de reducción de la dimensionalidad mediante algoritmo predictivo.

#### Reducción de la dimensionalidad mediante Ridge y Lasso

```
# Preparación de los datos.
datos <- stats::na.omit(health)
# Sacamos las columnas categóricas: Country, Region y Climate
columnas <- c('Health_exp_pct_GDP_2016','Health_exp_public_pct_2016','Health_exp_out_of_pocket_pct_2016', 
    'Health_exp_per_capita_USD_2016','per_capita_exp_PPP_2016','External_health_exp_pct_2016', 
    'Physicians_per_1000_2009.18','Nurse_midwife_per_1000_2009.18','Specialist_surgical_per_1000_2008.18',
    'Completeness_of_birth_reg_2009.18', 'Completeness_of_death_reg_2008.16','Population',
    'Area..sq..mi..', 'Pop..Density..per.sq..mi..','Coastline..coast.area.ratio.','Net.migration','Infant.mortality..per.1000.births.',
    'GDP....per.capita.','Literacy....','Phones..per.1000.','Arable....','Crops....','Other....',
    'Birthrate', 'Deathrate', 'Agriculture', 'Industry', 'Service', 'Total.cases','Total','Total.recovered',
    'Active.cases','Serious...Critical','Total.cases...1M.pop','Total.tests','Tests...1M.pop')

x <- stats::model.matrix(`Deaths...1M.pop` ~., data=datos)[ , columnas]
y <- datos$`Deaths...1M.pop`
```

```
if (comprobar_paquete("glmnet")) {library(glmnet)}
#library(glmnet)

set.seed(12345)

# Calculamos el valor de lambda mediante cross validation, 10 folds.
# MSE es el estimador utilizado para el cálculo del error.

# glmnet: para rigde alpha=0; para lasso alpha=1

reduccion <- function(alpha){
    if (alpha == 0) {
        nombre_modelo='RIDGE'
        #print("Reducción mediante RIDGE")
        #print("========================")
    }
    else {
        nombre_modelo='LASSO'
        #print("Reducción mediante LASSO")
        #print("========================")
    }
    modelo <- cv.glmnet(x, y, alpha=alpha, nfolds = 10, type.measure = "mse")
    main <- paste('Distribución del error, modelo ', nombre_modelo)
    plot(modelo, sub=main,  cex.sub=1.5)

    # Valor lambda con el que se consigue el mínimo test-error
    print(paste0(nombre_modelo, ". Mín. Error lamda: ", modelo$lambda.min))
    # Valor de lambda óptimo. Aquel que está lo más próximo a 1 desv. estándar del mínimo
    print(paste0(nombre_modelo, ". Valor óptimo de lamda: ", modelo$lambda.1se))

    modelo_final <- glmnet(x, y, alpha=alpha, lambda = modelo$lambda.1se )
    return(coef(modelo_final))
}

for (alpha in c(0,1)) {
    print(reduccion(alpha))
}
```

No se consigue una reducción de la dimensionalidad automática mediante Lasso o Ridge óptima. Ello puede ser debido a la nula correlación que hay entre las variables. Por ello lo más recomendable es utilizar la matriz de correlación para detectar las dimensiones que aportan más información visualmente.

## 4.2. Comprobación de normalidad y homogeneidad de la varianza

Para comprobación de la normalidad podemos utilizar el test Shapiro–Wilk

Para la comprobación de la homocedasticidad podemos utilizar el test de Levene


### Test Shapiro–Wilk
Vamos a definir el contraste de hipótesis siguiente:
* $H_0$: los datos proceden de una distribución normal
* $H_1$: los datos no proceden de una distribución normal

```
# Test Shapiro-Wilk
for (i in c("Phones_per_1k","GDP_per_capita", "Population", "Agriculture", "Birthrate", 
"Infant_mortality_1k", "Birth_regs", "Surgical_per_1k","Nurse_per_1k", "Physicians_per_1k", 
"Exp_per_capita", "Health_Exp_per_capita", "Health_exp_public_pct", "External_health_exp_pct")) {
  s_p <- shapiro.test(covidTrain[,i])$p.value
  cat("Valor p de",i,"=", s_p,"\n")
}
```

El valor $p$ es inferior a 0.05, por lo que podemos rechazar la hipótesis nula, es decir: las muestras no siguen una distribución normal.


### Test de Levene
El conjunto de datos no sigue una distribución normal, por lo que para el test de Levene utilizaremos la mediana como centro.

Vamos a definir el contraste de hipótesis siguiente:
* $H_0$: la varianza entre los grupos de muertes por Covid-19 es igual
* $H_1$: la varianza entre los grupos es distinta



```
# Convertimos a Factor la clusterizacion de muertes
covidTrain$Deaths_1m_clust <- as.factor(covidTrain$Deaths_1m_clust)
# Test de Levene
library(car)
for (i in c("Phones_per_1k","GDP_per_capita", "Population", "Agriculture", "Birthrate", 
"Infant_mortality_1k", "Birth_regs", "Surgical_per_1k","Nurse_per_1k", "Physicians_per_1k", 
"Exp_per_capita", "Health_Exp_per_capita", "Health_exp_public_pct", "External_health_exp_pct")) {
  l_p <- leveneTest(covidTrain[,i] ~ Deaths_1m_clust, data = covidTrain, center=median)$`Pr(>F)`
  cat("Valor p de",i,"=", l_p,"\n")
}
```

Con $p>0.05$ la hipótesis nula $H_0$ sólo se cumple para *Population* y *Physicians_per_1k*, por lo que podemos decir que la varianza del número de médicos por 1000 habitantes, y de la población con las muertes por Covid-19 es la misma.

No sucede lo mismo con las otras variables, por lo que en este caso rechazamos la hipótesis nula.


## 4.3. Aplicación de pruebas estadísticas para comparar los grupos de datos. En función de los datos y el objetivo del estudio, aplicar pruebas de contraste de hipótesis, correlaciones, regresiones, etc. Aplicar al menos tres métodos de análisis diferentes.

### Contraste de hipótesis

Vamos a ejecutar el test de Kruskal-Wallis para determinar si existen diferencias en el número de muertes según la región:

```
# Ejecución del test de Kruskal-Wallis
for (i in c("Region")) {
  k_p <- kruskal.test(Deaths_1m ~ covidTrain[,i], data = covidTrain)$p.value
  cat("Valor p de",i,"=", k_p,"\n")
}
```

La hipótesis nula $H_0$ es que no existen diferencias en el número de muertes/millón de habitantes en función de la región. La hipótesis alternativa $H_1$ es que al menos una de las poblaciones es distinta:
* $H_0: \mu_1=\mu_2=\mu_3=\mu_4 =\ ...$
* $H_1: \exists\mu_j\not=\mu\ \ \ \ j=1,2,3,4,\ ...$

Al ejecutar el test de Kruskal-Wallis, el valor p obtenido $p=0$, por lo que, tomando el valor habitual $\alpha=0,05$, se rechazaría la hipótesis nula: podemos afirmar que al menos una de las regiones es distinta de las demás con respecto al número de muertes por Covid-19.

# 5. Modelo
Vamos a utilizar Gradient Boosting. Lo probaremos para análisis de regresión y de clasificación, haciendo uso de las variables etiqueta *Deaths_1m* y *Deaths_1m_clust*.

## Análisis de regresión

```
# Generamos el modelo
covid.boost = gbm(Deaths_1m ~ ., data=covidTrain[c("Region","Phones_per_1k","GDP_per_capita", "Population", "Agriculture", "Birthrate", 
"Infant_mortality_1k", "Birth_regs", "Surgical_per_1k","Nurse_per_1k", "Physicians_per_1k", 
"Exp_per_capita", "Health_Exp_per_capita", "Health_exp_public_pct", "External_health_exp_pct","Deaths_1m")], n.trees = 10000)

# Información
summary(covid.boost)

# Predecimos utilizando los mismos datos de train
predTrain <- predict.gbm(covid.boost, covidTrain[c("Region","Phones_per_1k","GDP_per_capita", "Population", "Agriculture", "Birthrate", 
"Infant_mortality_1k", "Birth_regs", "Surgical_per_1k","Nurse_per_1k", "Physicians_per_1k", 
"Exp_per_capita", "Health_Exp_per_capita", "Health_exp_public_pct", "External_health_exp_pct")], n.trees=10000)
predTrain <- replace(predTrain, which (predTrain < 0), 0)

# Predecimos los valores de test, aunque no tendremos posibilidad de validarlos
predTest <- predict.gbm(covid.boost, covidTest[c("Region","Phones_per_1k","GDP_per_capita", "Population", "Agriculture", "Birthrate", 
"Infant_mortality_1k", "Birth_regs", "Surgical_per_1k","Nurse_per_1k", "Physicians_per_1k", 
"Exp_per_capita", "Health_Exp_per_capita", "Health_exp_public_pct", "External_health_exp_pct")], n.trees=10000)
predTest <- replace(predTest, which (predTest < 0), 0)
```

```
plot = ggplot(covidTrain, aes(x=Country)) + geom_line(aes(y=Deaths_1m), group=1, colour="red") + 
geom_line(aes(y=predTrain), group=1) + theme(axis.text.x = element_text(angle = 90, hjust = 1, size=0)) +
ggtitle("Muertes reales vs predicción del modelo (en rojo) datos de entrenamiento") + 
theme(plot.title = element_text(lineheight=.8, face="bold")) 

result = data.frame(covidTest$Country, predTest)

print(plot)
print("Predicción de muertes por millón para los países sin datos:")
print(result)
```

## Análisis de clasificación

```
# Modelo
covidTrain[, "Deaths_1m_clust"] <- as.factor(covidTrain[, "Deaths_1m_clust"])
covid.boost = gbm(Deaths_1m_clust ~ ., data=covidTrain[c("Region","Phones_per_1k","GDP_per_capita", "Population", "Agriculture", "Birthrate", 
"Infant_mortality_1k", "Birth_regs", "Surgical_per_1k","Nurse_per_1k", "Physicians_per_1k", 
"Exp_per_capita", "Health_Exp_per_capita", "Health_exp_public_pct", "External_health_exp_pct","Deaths_1m_clust")], n.trees = 10000)

# Información del modelo
summary(covid.boost)

# Predicción con datos de entrenamiento
PredTrain <- predict.gbm(covid.boost, covidTrain[c("Region","Phones_per_1k","GDP_per_capita", "Population", "Agriculture", "Birthrate", 
"Infant_mortality_1k", "Birth_regs", "Surgical_per_1k","Nurse_per_1k", "Physicians_per_1k", 
"Exp_per_capita", "Health_Exp_per_capita", "Health_exp_public_pct", "External_health_exp_pct")], n.trees=10000, type="response")
```

```
# Resultado de la predicción
labels = colnames(PredTrain)[apply(PredTrain, 1, which.max)]
result = data.frame(covidTrain$Deaths_1m_clust, labels)
print("Matriz de confusión de predicción con los datos de entrenamiento:")
table(result)

# Resultado de la predicción con datos de test
PredTest <- predict.gbm(covid.boost, covidTest[c("Region","Phones_per_1k","GDP_per_capita", "Population", "Agriculture", "Birthrate", 
"Infant_mortality_1k", "Birth_regs", "Surgical_per_1k","Nurse_per_1k", "Physicians_per_1k", 
"Exp_per_capita", "Health_Exp_per_capita", "Health_exp_public_pct", "External_health_exp_pct")], n.trees=10000, type="response")
labels = colnames(PredTest)[apply(PredTest, 1, which.max)]
result= data.frame(covidTest$Country, labels)
print("Predicción del grupo según muertes para los países sin datos:")
print(result)
```

La prueba del modelo de regresión sobre los mismos datos de entrenamiento obtiene un rendimiento excelente, a pesar de la poca información que dichos datos parecen aportar. A nivel gráfico se observa que ambas líneas se superponen en prácticamente todos los paises. 

Al hacer la prueba del modelo de clasificación, el resultado es similar, se obtiene un 99,4% de tasa de acierto. 

Con respecto a los datos de test, no disponemos de los datos reales, pero se sigue la misma tendencia en la que los países más desarrollados itenen más riesgo de muertes que los menos desarrollados.
